# Data

```python
files = [x for x in os.listdir("submissions/") if x.startswith("Submission")]
files
```

```python
preds = [pd.read_csv(f) for f in files]
```

```python
preds[5].iloc[5]
```

```python
pd.concat(preds).to_csv("test.csv", index=False)
```

```python
pd.concat([pred.Title.apply(len) for pred in preds], axis=1).apply(np.argmax, axis=1).value_counts()
```

```python
indices = pd.concat([pred.Title.apply(len) for pred in preds], axis=1).apply(np.argmax, axis=1)
indices.values
```

```python
preds_titles = pd.concat([pred.Title for pred in preds], axis=1)
preds_titles.columns = [f"Title{i}" for i in range(len(preds_titles.columns))]
```

```python
titles = []
for i, ind in enumerate(indices.values):
    titles.append(preds_titles.iloc[i][f"Title{ind}"])
```

```python
pd.DataFrame({
    "ID": preds[0].ID.values,
    "Title1": titles
}).to_csv("sub.csv", index=False)
```
